```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("GDIM_2023_03.csv")

# Filter for India
india = df[df['country'] == 'India']

# Only use MAX parental education and all children
india = india[(india['parent'] == 'max') & (india['child'] == 'all')]

# Sort by cohort
india = india.sort_values(by='cohort')
```

```python
india
```

```python
# --- 1. Plot: Mean years of schooling of parents vs children over cohorts ---
plt.figure(figsize=(10, 6))
plt.plot(india['cohort'], india['MEANp'], label="Parents", marker='o')
plt.plot(india['cohort'], india['MEANc'], label="Children", marker='s')
plt.title("Mean Years of Schooling in India by Cohort")
plt.xlabel("Birth Cohort")
plt.ylabel("Years of Schooling")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- 2. Plot: Probability child surpasses parent's education (CAT) ---
plt.figure(figsize=(10, 6))
plt.plot(india['cohort'], india['CAT'], marker='o', color='darkgreen')
plt.title("Probability of Children Surpassing Parents' Education (India)")
plt.xlabel("Birth Cohort")
plt.ylabel("Probability")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- 3. Plot: Educational Distribution of Children (C1 to C5) ---
plt.figure(figsize=(10, 6))
for level, label in zip(['C1', 'C2', 'C3', 'C4', 'C5'],
                        ['Less than Primary', 'Primary', 'Lower Secondary', 'Upper Secondary', 'Tertiary']):
    plt.plot(india['cohort'], india[level], label=label)

plt.title("Education Distribution of Children in India by Cohort")
plt.xlabel("Birth Cohort")
plt.ylabel("Share of Children")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- 4. Correlation & Beta Coefficient (persistence) ---
plt.figure(figsize=(10, 6))
plt.plot(india['cohort'], india['COR'], label="Correlation (COR)", marker='o')
plt.plot(india['cohort'], india['BETA'], label="Beta Coefficient", marker='x')
plt.title("Intergenerational Education Persistence in India")
plt.xlabel("Birth Cohort")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
